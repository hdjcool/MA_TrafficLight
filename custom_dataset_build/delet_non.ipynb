### non-emergency 라벨인 데이터 전량 삭제

```python
def remove_labels_with_class(labels_dir, images_dir, target_class=1):
    for label_file in os.listdir(labels_dir):
        label_path = os.path.join(labels_dir, label_file)
        
        if not os.path.isfile(label_path):
            continue
            
        with open(label_path, 'r') as f:
            lines = f.readlines()
            
        should_delete = any(line.startswith(f"{target_class} ") for line in lines)
        
        if should_delete:
            image_file = label_file.replace('.txt', '.jpg')
            image_path = os.path.join(images_dir, image_file)
            
            os.remove(label_path)
            if os.path.exists(image_path):
                os.remove(image_path)
```

```python
remove_labels_with_class(train_labels_dir, train_images_dir)
```

```python
remove_labels_with_class(val_labels_dir, val_images_dir)
```

```python
remove_labels_with_class(test_labels_dir, test_images_dir)
```

### 라벨 index number 변경

```python
base_dir = '/ultralytics/examples/Work_dir/urban_detection-3'
```

```python
class_mapping = {
    '0': '1',
    '1': '4',
    '2': '5',
    '3': '2'
}
```

```python
def update_label_file(file_path, class_mapping):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    updated_lines = []
    for line in lines:
        parts = line.strip().split()
        if parts:  # 비어있는 줄이 아닌 경우 처리
            class_id = parts[0]
            # 클래스 ID를 매핑 테이블에 따라 변환
            if class_id in class_mapping:
                print('yes')
                parts[0] = class_mapping[class_id]
            updated_lines.append(' '.join(parts))
    
    # 원본 파일에 덮어쓰기
    with open(file_path, 'w') as f:
        f.write('\n'.join(updated_lines) + '\n')
```

```python
for folder in ['test']:
    labels_dir = os.path.join(base_dir, folder, 'labels')
    if os.path.exists(labels_dir):
        for file_name in os.listdir(labels_dir):
            if file_name.endswith('.txt'):
                file_path = os.path.join(labels_dir, file_name)
                update_label_file(file_path, class_mapping)
```

### Custom dataset으로 각 데이터 copy

```python
base_dir = '/ultralytics/examples/Work_dir/custom_dataset'
```

```python
for big_folder in ['kitti_dataset', 'emergency-2', 'emergency-nonemergency-vehicle-7']:
    for task in ['train', 'valid']:
        for datas in ['images', 'labels']:
            target_dir = os.path.join(base_dir , task, datas)
            source_dir = os.path.join('/ultralytics/examples/Work_dir', big_folder, task, datas)
            for file_name in os.listdir(source_dir):
                if file_name.endswith('.jpg') or file_name.endswith('.txt'):
                    source_path = os.path.join(source_dir, file_name)
                    target_path = os.path.join(target_dir, file_name)
                    shutil.copy(source_path, target_path)
```
